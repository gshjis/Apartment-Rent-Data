## 📝 Анализ текстовых описаний (LDA)

Текстовое поле `body` содержит детальные описания объектов аренды. Для извлечения скрытых тем и их связи с ценой применяется **Latent Dirichlet Allocation (LDA)** — генеративная вероятностная модель, которая представляет каждый документ как смесь тем, а каждую тему — как распределение слов.

### Цели анализа:
1. Выявить основные темы, обсуждаемые в объявлениях
2. Оценить корреляцию каждой темы с ценой аренды
3. Использовать вероятности тем как дополнительные признаки для модели

### Параметры модели:
- Количество тем: **15**
- Alpha (разреженность тем в документах): **0.01**
- Eta (разреженность слов в темах): **0.01**
- Метод корреляции: **Spearman** (устойчив к нелинейным зависимостям)
- Проверка значимости: **Permutation test** (100 перестановок)

### Предобработка текста:
- Приведение к нижнему регистру
- Удаление стоп-слов (nltk.corpus.stopwords)
- Удаление пунктуации и цифр
- Лемматизация (WordNetLemmatizer)
- Фильтрация слов по частоте (min_df=10, max_df=0.5)

```python
from analysis import LdaTopicModel
LDA_model = LdaTopicModel(
    dataset=train, 
    text_column="body", 
    n_topics=15,
    alpha=0.01,
    eta=0.01
)
LDA_model.fit()

# Корреляция вероятностей тем с ценой
corr_df = LDA_model.topics_price_correlation(
    price_column="price",
    correlation_method="spearman",  # или "pearson"
    significance_test=True,
    n_permutations=100
)
```

```python
import matplotlib.pyplot as plt

# Построить график корреляций
plt.figure(figsize=(12, 6))
plt.bar(corr_df['topic'], corr_df['corr'])
plt.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
plt.xlabel('Topic')
plt.ylabel('Correlation with price')
plt.title('Topic-Price Correlations')
plt.show()
```

## 🧪 Экспериментальный дизайн и гипотезы

### Baseline модель

В качестве базового решения (`baseline`) будет использована **единая модель XGBoost**, обученная на всех данных без разделения по географическим группам. Это позволит оценить, насколько более сложные подходы превосходят простую глобальную модель.

**Признаки baseline:**
- Числовые: `bedrooms`, `square_feet`, `bathrooms`
- Категориальные: `state`, `category`, `fee`, `source` (One-Hot Encoding)
- Географический: `is_anomaly` (бинарный признак: 1 — экстремальный город, 0 — обычный)
- Текстовые: вероятности тем LDA (15 признаков)

---

### Экспериментальные гипотезы

#### Гипотеза 1: Разделение моделей по географическим группам повышает качество

| Группа | Подход | Обоснование |
|--------|--------|-------------|
| **Обычные города** | Отдельная модель | Данные однородны, структура цен предсказуема |
| **Экстремальные города** | Отдельная модель | Цены значимо отличаются от среднего по штату, требуют отдельного подхода |

**Экстремальные города** — города, определённые в предыдущем анализе как статистически значимо отличающиеся от среднего по штату (p < 0.05, отклонение медианы > 30%). В рамках сегментированного подхода для экстремальных городов дополнительно добавляется признак `anomaly_type` (premium / budget), указывающий на направление отклонения.

Для каждой группы будет обучена своя модель XGBoost.

---

#### Гипотеза 2: Способ кодирования текстовых признаков влияет на качество

Будут протестированы два подхода к работе с текстовыми описаниями:

| Метод | Описание | Преимущества | Недостатки |
|-------|----------|--------------|------------|
| **LDA (вероятности тем)** | Каждый документ представлен распределением по 15 темам | Интерпретируемость, снижение размерности, учёт контекста | Потеря конкретных терминов |
| **One-Hot Encoding (топ-N слов)** | Бинарные признаки для N наиболее частотных слов | Сохранение конкретных терминов, простота | Высокая размерность, разреженность, потеря контекста |

Для OHE будет отобрано **топ-100 слов** по частоте (после удаления стоп-слов и лемматизации).

**Дополнительно:** на основе LDA будет построен агрегированный признак — **средневзвешенная корреляция тем с ценой**. Для каждого объявления вычисляется:

$$
\text{topic\_weighted\_correlation} = \sum_{i=1}^{n} p_i \times \text{r}_i
$$

где:
- $p_i$ — вероятность i-й темы, 
- $\text{r}_i$ — корреляция i-й темы с ценой. 

Этот признак аккумулирует информацию о тональности описания в одной числовой переменной.

---

#### Гипотеза 3: CatBoost может дать преимущество за счёт встроенной обработки категориальных признаков

Для сравнения архитектур будет обучен **CatBoost** — градиентный бустинг, который эффективно работает с категориальными признаками без необходимости OHE.

Сравниваются две конфигурации CatBoost:
- **CatBoost (единая модель)** — аналогично XGBoost baseline, с теми же признаками
- **CatBoost (сегментация)** — отдельные модели для обычных и экстремальных городов

CatBoost использует нативные категориальные признаки (`state`, `category`, `fee`, `source`, `anomaly_type`), что потенциально может улучшить качество и упростить пайплайн.

### Схема эксперимента

**1. Подготовка данных**

- Разделение train/test
- Формирование признакового пространства:
  - Базовые признаки (числовые + категориальные)
  - Географический признак (is_anomaly, а в сегментированных моделях — anomaly_type)
  - Текстовые признаки (2 варианта кодирования + агрегированный признак)

**2. Эксперименты**

| № | Название | Конфигурация |
|---|----------|--------------|
| **A** | XGBoost baseline | Единая модель, LDA-признаки, is_anomaly |
| **B1** | XGBoost гео-сегментация (LDA) | 2 модели (обычные / экстремальные города), LDA-признаки, anomaly_type для экстремальных |
| **B2** | XGBoost гео-сегментация (OHE) | 2 модели, OHE (топ-100), anomaly_type для экстремальных |
| **B3** | XGBoost гео-сегментация (LDA + агрегат) | 2 модели, LDA-признаки + средневзвешенная корреляция, anomaly_type для экстремальных |
| **C1** | CatBoost baseline | Единая модель, LDA-признаки, is_anomaly (категориальные признаки нативные) |
| **C2** | CatBoost гео-сегментация | 2 модели, LDA-признаки, anomaly_type для экстремальных (категориальные признаки нативные) |

**3. Валидация**

- **Кросс-валидация:** GroupKFold по городам (чтобы объекты одного города не разбегались по разным фолдам)
- **Метрики:** RMSE, MAE, MAPE

**5. Интерпретация**

- Сравнение **A vs B1–B3** покажет, даёт ли сегментация по географии выигрыш в XGBoost
- Сравнение **B1 vs B2** покажет, какой способ кодирования текстов (LDA или OHE) оптимален
- Сравнение **B1 vs B3** покажет, улучшает ли качество добавление агрегированного признака
- Сравнение **A vs C1** покажет, даёт ли CatBoost преимущество на той же архитектуре
